```python
# Name - Akshanshu Bharti 
# Task -2 (Image to Pencil Sketch with Python:)We need to read the image in RBG format and then convert it to a grayscale image.
This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale 
image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. 
Then we can finally create the pencil sketch by mixing the grayscaleimage with the inverted blurry image. 
This can be done by dividing the grayscale image by the inverted blurry image. 
Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.
```

```python
Step 1 : I am Import the Photo or Image in the notebook (I used the Thar Car Image.)
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("thar.jpg")
cv2.imshow("Thar", image)
cv2.waitKey(0)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("New Thar", gray_image)
cv2.waitKey(0)
```

```python
pic = cv2.imread("thar.jpg")
plt.axis(False)
plt.imshow(pic)
plt.show()
```

```python
# Now step 2: Converting original image into RGB Format 
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("New Thar", gray_image)
cv2.waitKey(0)
pic_1 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(pic_1)
plt.axis(False)
plt.show()
```

```python
# Step 3 Converting Original image into grayscale
```

```python
pic_2 = cv2.cvtColor(pic_1, cv2.COLOR_BGR2GRAY)
plt.imshow(pic_2)
plt.axis(False)
plt.show()
```

```python
# step4 converted image into inverted 
```

```python
inverted_image = 255 - gray_image
cv2.imshow("Inverted", inverted_image)
cv2.waitKey()
plt.imshow(inverted_image)
plt.axis(False)
plt.imshow(inverted_image)
plt.show()
```

```python
#Step 5 Now I am converted inverted image into  inverting blurring format
```

```python
inverted_blurring = cv2.GaussianBlur(inverted_image, (111,111), sigmaX=0, sigmaY=0)
plt.imshow(inverted_blurring)
plt.axis(False)
plt.show()
```

```python
#Step 6 Converting image into blurred format
```

```python
blurred_format = cv2.bitwise_not(inverted_blurring)
plt.imshow(blurred_format)
plt.axis(False)
plt.show()
```

```python
#Step 7 Mixing by dividing inverted blurry image and greyscale format image
```

```python
pic_7 = cv2.divide(pic_2,blurred_format , scale=250.0)
plt.imshow(pic_7)
plt.axis(False)
plt.show()
```

```python
# The last step is  Converting into the pencil sketch format by using last mixup image
```

```python
plt.figure(figsize = (12,8))
pic_8 = cv2.cvtColor(pic_7, cv2.COLOR_RGB2BGR)
plt.imshow(pic_8)
plt.axis(False)
plt.show()
```
